```python
#IG
import numpy as np
from math import log2
#target
y=np.array([0,0,1,1])
def entropy(y):
  classes=np.unique(y)
  total=len(y)
  ent=0
  for c in classes:
    p=np.sum(y==c)/total
    ent-=p*log2(p)
  return ent
  #entropy before split
parent_entropy=entropy(y)
  #suppose we split at hours <=4
left=y[:2]
right=y[2:]
  #weighted entropy after split
n=len(y)
weighted_entropy=(len(left)/n)*entropy(left)+(len(right)/n)*entropy(right)
  #IG
IG=parent_entropy-weighted_entropy
print("Parent entropy:",parent_entropy)
print("Weighted entropy:",weighted_entropy)
print("Information Gain:",IG)
```
